```python
idadesPnad = dadosPnad['V2009'].unique()
idadesPnad.sort()
idadesPnad
```

```python
idades = [0] * 113
for i in range(113):
  idadesList[i] = i
```

```python
salariosIdade = [0] * 113
soma = 0
for i in range(113):
  soma = 0
  print('indice', i)
  for dado in dadosPnad.index :
    if (dadosPnad['V2009'][dado] == i):  
      if float((dadosPnad['VD4020'][dado]) > 0):             
        if (soma == 0):
          soma += float(dadosPnad['VD4020'][dado])             
        else:
          soma = (soma + float(dadosPnad['VD4020'][dado])) / 2
        salariosIdade[i] = soma          
```

```python
plt.title('Média de rendimentos mensais por idade')
plt.xlabel('Idade')
plt.ylabel('Rendimento médio mensal em R$')
plt.plot(idadesList, salariosIdade)
```
